#### 1. Clean the data. Use .apply and lambda functions where you can

```python
# rename the columns to remove spaces and extra characters
def clean_column_spaces(frameCols):
    
    colList = [x for x in frameCols]
    trim_list = [x.strip() for x in frameCols]

    col_tup = zip(colList, trim_list) 
    return col_tup
    

print df.columns
col_tup = clean_column_spaces(df)
do = [df.rename(columns={col_tup[x][0]: col_tup[x][1]}, inplace=True) for x in range(len(col_tup))]

print df.columns
```

```python
# convert the funding, year founded, and funding_rounds columns to integers
print df.dtypes
#df[list("ABCD")].astype(int)
intList = ['funding_total_usd', 'founded_year', 'funding_rounds']
print df['funding_total_usd'].unique()

df.fillna(0,inplace=1)

def clean_numbers(value):
    #print value
    value = value.strip()
    value = value.replace(',','0')
    value = value.replace("-",'0')
    return value

if df['funding_total_usd'].dtype != 'int64':
    df['funding_total_usd'] = df['funding_total_usd'].apply(clean_numbers)
    df['funding_total_usd'] = df['funding_total_usd'].astype(int)

if df['founded_year'].dtype != 'int64':
    df['founded_year'] = df['founded_year'].astype(int)

    
if df['funding_rounds'].dtype != 'int64':
    df['funding_rounds'] = df['funding_rounds'].astype(int)

print df.dtypes
df.head(3)
#print df['funding_total_usd']
```

```python
# convert category_list so that each cell is a list of categories
def split_pipes(value):
    #print type(value)
    if value != 0:
        value = value.strip('|')
        value = value.split('|')
        return value

df['category_list']  = df['category_list'].apply(split_pipes)
df
```

```python
# use value_counts on market, status, and country_code to check for any repeats or misnamed entries
print df['market'].value_counts().to_frame().sort()
print df['status'].value_counts().to_frame().sort()
print df['country_code'].value_counts().to_frame().sort()
```

```python
# BONUS: convert the date columns to appropriate datetime objects
head = df
#print head[head['last_funding_at'].isnull()]
def count_len(value):
    if len(value) > 8:
        value = value.replace('1000','2000')
        #print value
        return value

head['first_funding_at'] = head['first_funding_at'].apply(count_len)
head['last_funding_at'] = head['last_funding_at'].apply(count_len)
#print head['last_funding_at'].unique()
head['first_funding_at'] = pd.to_datetime(head['first_funding_at'])
head['last_funding_at'] = pd.to_datetime(head['last_funding_at'])
head.dtypes
```
